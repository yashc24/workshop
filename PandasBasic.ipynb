# ***Import Pandas***

```python
import pandas as pd
import numpy as np
```

## ***Series***

```python
country = ["india","pakistan","USA","Nepal","srilanla"]
pd.Series(country)
```

```python
food = ["momos","pizza","pav bhaji","burger","pasta"]
fav = pd.Series(food)
fav
```

```python
#custom index
marks = [78,90,67,99]
sub = ["maths","science",'english',"marathi"]
pd.Series(marks,index=sub)
```

```python
# name
marks = [78,90,67,99]
sub = ["maths","science",'english',"marathi"]
pd.Series(marks,index=sub,name="student marksheet")
```

## ***Creating Series Object from Dictionary***

```python
# dictionary
marks ={
    "maths":78,
    "science":90,
    "english":67,
    "marathi":99
}
pd.Series(marks)
```

```python
marks ={
    "maths":78,
    "science":90,
    "english":np.nan,
    "marathi":99
}
pd.Series(marks,name="student marksheet")
```
